```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({"text.usetex": True, "font.family": "Arial"})
```

```python
num_runs = 20
eq_t = 100
shape = np.load("../outputs/brownian_motion_01.npz")["pos"][eq_t:,-1,:2].shape
pos = np.zeros((shape[0], num_runs, 2))
for i in range(num_runs):
    pos[:,i,:] = np.load(f"../outputs/brownian_motion_{i+1:02d}.npz")["pos"][eq_t:,-1,:2]
```

```python
fig, ax = plt.subplots()
ax.set_title("Brownian motion", fontsize=20)
ax.set_xlabel(r"$x$", fontsize=15)
ax.set_ylabel(r"$y$", fontsize=15, rotation=0)
ax.plot(pos[:,:,0], pos[:,:,1])
```

```python
D_vec = np.diff(pos, axis=0)
D = np.linalg.norm(D_vec, axis=2)
SD = np.square(D)
MSD = np.mean(SD, axis=1)
t = np.arange(0, 4.0, 0.01)
```

```python
fig, ax = plt.subplots()
ax.set_title("Brownian motion, MSD vs time", fontsize=20)
ax.set_xlabel(r"$t$", fontsize=15)
ax.set_ylabel(r"$MSD$", fontsize=15, rotation=0)
ax.plot(t[:-1], MSD)
```
